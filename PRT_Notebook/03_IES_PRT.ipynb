```python
import os
import shutil
import subprocess
import matplotlib.pyplot as plt;
import psutil
import pyemu
import flopy
import socket
import warnings
warnings.filterwarnings(action='ignore')
import pandas as pd
import geopandas as gpd
import numpy as np
import pathlib as pl
import shapely
from IPython.display import display
import math
import time
```

```python
cwd = os.getcwd()
scenarios_dir = pl.Path('..','Scenario_Notebook').resolve()
print('cwd:',cwd)
print('scenarios:',scenarios_dir)
```

```python
# Get list of all folders in base_scenario_notebook folder
folders = [f.name for f in os.scandir(scenarios_dir) if f.is_dir()]
folders
```

```python
# Filter to just our possible scenarios
scenario_folders = []
for file in folders:
    if file.startswith('R25'):
        scenario_folders.append(file)
for file in folders:
    if file.startswith('C25'):
        scenario_folders.append(file)
for file in folders:
    if file.startswith('R100'):
        scenario_folders.append(file)
for file in folders:
    if file.startswith('C100'):
        scenario_folders.append(file)
print(scenario_folders)
```

```python
for scenario in scenario_folders:
    # copy over completed IES run into temp folder
    subprocess.call(['robocopy', str(scenarios_dir) +'\\'+ str(scenario), str(cwd) + '\\temp', '/E'])

    temp_dir = os.path.join(cwd,'temp')
    
    # Make folder that final shps will be created in
    if os.path.exists(os.path.join(cwd,scenario)):
        shutil.rmtree(os.path.join(cwd,scenario))
    os.makedirs(os.path.join(cwd,scenario))

    # Get model_root_name from listing file
    for file in os.listdir(temp_dir):
        if file.endswith(".dis"):
            disfile = file
    gwfmodel_root_name = os.path.splitext(disfile)[0]
    print(gwfmodel_root_name)

    # Load model
    sim = flopy.mf6.MFSimulation.load(f'{gwfmodel_root_name}', sim_ws=temp_dir, exe_name='mf6.exe')
    sim.set_sim_path(temp_dir)
    model = sim.get_model(f'{gwfmodel_root_name}')

    # Get geodataframe and nodes of model for use later
    modelprops = model.modelgrid.geo_dataframe.reset_index()
    modelprops = modelprops.rename(columns={'index':'node'})
    # Make 1 based
    modelprops['node'] +=1

    # Make prt model
    prt_root_name = "prt_model"
    print(prt_root_name)
    
    # Create prt model
    prt = flopy.mf6.ModflowPrt(sim, modelname=prt_root_name, model_nam_file=f'{prt_root_name}.nam')
    
    # Create dis from base model
    prtdis = flopy.mf6.ModflowGwfdis(prt, length_units=model.modelgrid.units, xorigin=model.modelgrid.xoffset,
                                  yorigin=model.modelgrid.yoffset, nlay=model.modelgrid.nlay, nrow=model.modelgrid.nrow,
                                  ncol=model.modelgrid.ncol, delc=model.modelgrid.delc, delr=model.modelgrid.delr,
                                  top=model.modelgrid.top, botm=model.modelgrid.botm, idomain=model.modelgrid.idomain,
                                  pname=prt_root_name+'.dis')
    
    # Get well locations, skipping first one, which is on east side of Riv
    wel_loc = []
    for i in range(1,len(model.wel.stress_period_data.data[0])):
        wel_loc.append(model.wel.stress_period_data.data[0][i][0])
    display(wel_loc)
    
    # Get Well Node Locations
    well_nodes = []
    for i in range(len(wel_loc)):
        well_nodes.append(model.modelgrid.get_node(wel_loc[i]))
    display(well_nodes)
    
    # Assign a different zone number to well cells.
    # This makes it easier to determine where particles terminate.
    
    # Make izone and all 0's
    izone = np.zeros((model.modelgrid.nlay, model.modelgrid.nrow, model.modelgrid.ncol), dtype=int)
    izone = np.ravel(izone)
    
    # Change well nodes to a 1
    for i in range(len(well_nodes)):
        node = well_nodes[i]
        izone[node] = 1
    
    # Remake into correct shape
    izone = izone.reshape(model.modelgrid.nlay, model.modelgrid.nrow, model.modelgrid.ncol)
    
    # Create PRT model input package
    mip = flopy.mf6.ModflowPrtmip(prt, pname='mip', porosity=0.1, izone=izone)
    
    # Load Particle starting locs csv
    starting_locs = pd.read_csv(os.path.join(cwd,'PRT_Starting_Locs.csv'))
    
    # subtract 1 from row and column
    starting_locs['row'] -=1
    starting_locs['column'] -=1
    
    starting_locs.reset_index(inplace=True)
    starting_locs.rename(columns={'index':'pid'}, inplace=True)
    
    # Add o in front of every unique number, turns to string
    starting_locs['node'] = 'nn' + starting_locs['node'].astype(str)
    
    # Add layer column
    starting_locs['lay'] =0
    starting_locs['z'] =1
    
    # Make prt_releasepts tuple
    starting_locs['prt_releasepts'] = list(zip(starting_locs['pid'], starting_locs['lay'], starting_locs['row'], starting_locs['column'],
                                               starting_locs['centerX'], starting_locs['centerY'], starting_locs['z'], starting_locs['node']))
    
    # Convert to list
    prt_releasepts = starting_locs['prt_releasepts'].tolist()
    
    prp = flopy.mf6.ModflowPrtprp(prt, pname='prp', filename=f'{prt_root_name}.prp', nreleasepts=len(starting_locs),
                                  packagedata=prt_releasepts, perioddata={0: ["FIRST"]}, exit_solve_tolerance=1e-5,
                                  extend_tracking=True, boundnames=True, istopzone=1, stoptime=9131.25, local_z=True)
    
    # Create output control package
    oc = flopy.mf6.ModflowPrtoc(prt, pname='oc',budget_filerecord=[f'{prt_root_name}.bud'],
                                trackcsv_filerecord=[f'{prt_root_name}.trk.csv'], track_release=True, track_terminate=True)
    
    # Flow model interface
    exg = flopy.mf6.ModflowGwfprt(sim, exgmnamea=gwfmodel_root_name, exgmnameb=prt_root_name, filename=f"{gwfmodel_root_name}.gwfprt")
    
    ems = flopy.mf6.ModflowEms(sim, pname='ems', filename=f'{prt_root_name}.ems')
    sim.register_solution_package(ems, [prt.name])
    
    sim.set_all_data_external()
    sim.write_simulation()

    # Locate parcsv files in directory
    parcsvfile = []
    for file in os.listdir(temp_dir):
        if file.endswith('par.csv'):
            parcsvfile.append(file)
    print(parcsvfile)
    
    # Locate iterations 1, 2, 3, 4 amongst those csv files
    csv123file = []
    for item in parcsvfile:
        if f'{gwfmodel_root_name}.0' in item:
            csv123file.append(item)
    for item in parcsvfile:
        if f'{gwfmodel_root_name}.1' in item:
            csv123file.append(item)
    for item in parcsvfile:
        if f'{gwfmodel_root_name}.2' in item:
            csv123file.append(item)
    for item in parcsvfile:
        if f'{gwfmodel_root_name}.3' in item:
            csv123file.append(item)
    for item in parcsvfile:
        if f'{gwfmodel_root_name}.4' in item:
            csv123file.append(item)
    print(csv123file)

    # check to see what all iterations were completed and make list to loop through
    for parcsv in csv123file:
        for item in parcsv:
            if '0' in item:
                iteration = '0'
        for item in parcsv:
            if '1' in item:
                iteration = '1'
        for item in parcsv:
            if '2' in item:
                iteration = '2'
        for item in parcsv:
            if '3' in item:
                iteration = '3'
        for item in parcsv:
            if '4' in item:
                iteration = '4'
        print(parcsv)
        print(iteration)
    
        # Locate pst file in directory
        pstfile = []
        for file in os.listdir(temp_dir):
            if file.endswith('.pst'):
                pstfile.append(file)
        print(pstfile)
        
        # Load pst file into pyemu
        pst = pyemu.Pst(os.path.join(temp_dir,pstfile[0]))
    
       # Load parfile csv
        parfilecsv = pd.read_csv(os.path.join(cwd,temp_dir,parcsv))
        parfilecsv = parfilecsv.rename(columns={parfilecsv.columns[0]: 'Realization'})
        realsparfile = parfilecsv['Realization'].to_list() 
    
        # Run parrep on pst files
        pstfiles = []
        for i in realsparfile:
            pst.parrep(parfile=os.path.join(temp_dir,parcsv), real_name=i, noptmax=0)
            pst.write(os.path.join(temp_dir,f'realizations_{i}.pst'), version=2)
            pstfiles.append(os.path.join(temp_dir,f'realizations_{i}.pst'))
        
        # Make work1 folder
        subprocess.call(["robocopy", temp_dir, temp_dir + '\work1', '/E'])
        if os.path.exists(os.path.join(temp_dir, 'work1','work1')):
            shutil.rmtree(os.path.join(temp_dir, 'work1','work1'))
        
        # Create pathline folders
        try:
            os.mkdir(os.path.join(temp_dir,f"Endpoints{iteration}"))
        except:
            pass
        
        #copy all files from Work1 to other work folders
        cores = 12
        counter = 2
        filedir = os.path.join (temp_dir, "work1")
        while counter <=cores:
            filename = "work" + str(counter)
            workfolder = os.path.join(temp_dir,filename)
            shutil.copytree(filedir,workfolder)
            counter += 1
        
        # Get chunks to split into
        chunksize = int(math.ceil(len(pstfiles) / cores))
        print(chunksize)
        pstfiles_split = [pstfiles[i * chunksize:(i+1) * chunksize] for i in range((len(pstfiles) + chunksize-1) // chunksize)]
        
        # Begin making outfiles
        counter = 1
        while counter <=cores:
            filename = os.path.join(temp_dir,"outfile" + str(counter) + ".py")
            f = open(filename, "w")
            f.write('import os, shutil, pyemu' + '\n')
            #f.write('cwd = os.getcwd()' + '\n')
            f.close()
            counter += 1 
        
        # Make outfile#.py files
        outfilecounter = 1
        for i in pstfiles_split:
            templist = i
            filename = os.path.join(temp_dir,'outfile' + str(outfilecounter) + '.py')
            for a in templist:
                pst_name = os.path.basename(a)
                prt_name = pst_name.lstrip('realizations_')
                prt_name = prt_name.rstrip('par.pst')
                f = open(filename, "a")
                f.write('try:\n')
                f.write(f'     pyemu.os_utils.run("pestpp-glm {pst_name}", cwd=r"{temp_dir}\work{outfilecounter}")'  + '\n')
                f.write(f'     shutil.move(r"{temp_dir}\work{outfilecounter}\{prt_root_name}.trk.csv",r"{temp_dir}\Endpoints{iteration}\prt_r{prt_name}.csv")' + '\n')
                f.write('except:\n')
                f.write('     pass\n')
                f.write(f'shutil.rmtree(r"{temp_dir}\work{outfilecounter}\{prt_root_name}.trk.csv", ignore_errors=True)' + '\n')
                f.close()
            outfilecounter +=1
        
        # Make run_outfiles.py
        filename = os.path.join(temp_dir,'run_outfiles.py')
        f = open(filename, 'w')
        f.write('import threading, subprocess, os' + '\n')
        f.write(f'temp_dir = r"{temp_dir}"' + '\n')
        f.write('def run_outfiles(outfile_name):' + '\n')
        f.write('     subprocess.run(["python", outfile_name], cwd=temp_dir)' + '\n')
        f.write('if __name__ == "__main__":' + '\n')
        for i in range(cores):
            i = i+1
            f.write(f'     outfile{i}_thread = threading.Thread(target=run_outfiles, args=("outfile{i}.py",))' + '\n')
        for i in range(cores):
            i = i+1
            f.write(f'     outfile{i}_thread.start()' + '\n')
        for i in range(cores):
            i = i+1 
            f.write(f'     outfile{i}_thread.join()' + '\n')
        f.close()
        
        # run outfiles
        %run "temp/run_outfiles.py"
        #run_magic_command = os.path.join(temp_dir,"run_outfiles.py")
        #%run $run_magic_command
    
        # Delete work folders
        counter = 1
        filedir = os.path.join (temp_dir, "work1")
        while counter <=cores:
            filename = "work" + str(counter)
            workfolder = os.path.join(temp_dir,filename)
            shutil.rmtree(workfolder)
            counter += 1 
    
        # Delete reals par files out of main directory
        for file in realsparfile:
            try:
                if os.path.isfile(file):
                    os.remove(file)
            except:
                pass
        
        endpoint_dir = os.path.join(temp_dir, f'Endpoints{iteration}')
        # Locate csv files in directory
        csvfiles = []
        for file in os.listdir(endpoint_dir):
            if file.endswith('.csv'):
                csvfiles.append(file)
        print(csvfiles)
    
        # Loop through prt csv files and get cells that end in izone of 1 (ie, were captured by a well)
        for i in csvfiles:
            mf6pl = pd.read_csv(os.path.join(endpoint_dir,i))
            # index by particle group and particle ID
            mf6pl.set_index(["iprp", "irpt"], drop=False, inplace=True)
            
            # determine which particles ended up in which capture zone
            mf6pl["destzone"] = mf6pl[mf6pl.istatus > 1].izone
            mf6pl["dest"] = mf6pl.apply(lambda row: ("well" if row.destzone == 1 else pd.NA), axis=1)
            
            mf6pl_in_well_df = mf6pl[mf6pl['dest']=='well']
            
            # Get starting izone = 0 to get the initial cell
            mf6pl_in_well_df = mf6pl_in_well_df[mf6pl_in_well_df['izone'] ==0]
            
            # Note, icell is the 1 based node number
            mf6pl_in_well_df.to_csv(os.path.join(endpoint_dir,f'{i}cap.csv'))   
    
        # Locate csv files ending in cap.csv
        csvfiles = []
        for file in os.listdir(endpoint_dir):
            if file.endswith('cap.csv'):
                csvfiles.append(file)
        print(csvfiles)      
    
        # Loop through cap.csv files and get icell values into a list
        node_occurrences = []
        for i in csvfiles:
            capture_df = pd.read_csv(os.path.join(endpoint_dir, i))
            unique_values = capture_df['icell'].unique().tolist()
            node_occurrences+=list(unique_values)
    
        # Get count of node, turn to dataframe
        node_counts_df = pd.Series(node_occurrences).value_counts().to_frame().reset_index()
        node_counts_df = node_counts_df.rename(columns={"index":"node"})
        
        # Calculate probability
        node_counts_df['prob'] = node_counts_df['count'] / int(len(csvfiles))
        node_counts_df['prob'] = node_counts_df['prob']*100
        display(node_counts_df)
    
        # Merge to get geometry
        node_counts_df = node_counts_df.merge(modelprops[['node','geometry']])
        
        # convert to geodataframe
        node_counts_df = gpd.GeoDataFrame(node_counts_df, geometry=node_counts_df['geometry'])
        display(node_counts_df.plot())
        
        node_counts_df.to_file(os.path.join(cwd, scenario, f'{scenario}_{iteration}.shp'))
        print(f'exported {scenario}{iteration}.shp')

        # Rejection of mean +- 3* stdev
        phiactualcsv = pd.read_csv(os.path.join(temp_dir,f'{gwfmodel_root_name}.phi.actual.csv'))
        phiactualcsv_std_low = phiactualcsv.at[int(iteration),'mean']-phiactualcsv.at[int(iteration),'standard_deviation']*3
        phiactualcsv_std_high = phiactualcsv.at[int(iteration),'mean']+phiactualcsv.at[int(iteration),'standard_deviation']*3
        phiactualcsv = phiactualcsv.drop(columns=['iteration','total_runs','mean','standard_deviation','min','max'])
        phiactualcsv_current_iter = phiactualcsv.iloc[int(iteration)].to_frame()
        filtered = phiactualcsv_current_iter.loc[(phiactualcsv_current_iter[int(iteration)]>=phiactualcsv_std_low) & (phiactualcsv_current_iter[int(iteration)]<=phiactualcsv_std_high)]
        filtered = filtered.index.tolist()
        filtered_csvfiles = ['prt_r' + element + '.csvcap.csv' for element in filtered]
        rejected = phiactualcsv_current_iter.loc[(phiactualcsv_current_iter[int(iteration)]<phiactualcsv_std_low) | (phiactualcsv_current_iter[int(iteration)]>phiactualcsv_std_high)]
        rejected  = rejected.index.tolist()
        with open(os.path.join(cwd, scenario, f'{scenario}_{iteration}_rejected.txt'), 'w') as file:
            file.write(str(rejected))

        # Loop through cap.csv files and get icell values into a list
        node_occurrences = []
        for i in filtered_csvfiles:
            if os.path.isfile(os.path.join(endpoint_dir, i)):
                capture_df = pd.read_csv(os.path.join(endpoint_dir, i))
                unique_values = capture_df['icell'].unique().tolist()
                node_occurrences+=list(unique_values)
    
        # Get count of node, turn to dataframe
        node_counts_df = pd.Series(node_occurrences).value_counts().to_frame().reset_index()
        node_counts_df = node_counts_df.rename(columns={"index":"node"})
        
        # Calculate probability
        node_counts_df['prob'] = node_counts_df['count'] / int(len(filtered_csvfiles))
        node_counts_df['prob'] = node_counts_df['prob']*100
        display(node_counts_df)
    
        # Merge to get geometry
        node_counts_df = node_counts_df.merge(modelprops[['node','geometry']])
        
        # convert to geodataframe
        node_counts_df = gpd.GeoDataFrame(node_counts_df, geometry=node_counts_df['geometry'])
        display(node_counts_df.plot())
        
        node_counts_df.to_file(os.path.join(cwd, scenario, f'{scenario}_{iteration}_rej.shp'))
        print(f'exported {scenario}{iteration}_rej.shp')


        
        # Delete endpoints folder
        if os.path.exists(os.path.join(temp_dir,f"Endpoints{iteration}")):
            shutil.rmtree(os.path.join(temp_dir,f"Endpoints{iteration}"))

    # Delete temp folder
    if os.path.exists(os.path.join(cwd,'temp')):
        shutil.rmtree(os.path.join(cwd,'temp'))
```
